```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
def standing_waves(N, tube):
    
    L = 10
    A = 1
    v = 1
    x = np.linspace(0,L,1000)
    
    if tube=='open':
        lambdaN = 2*L/N
        yp = A * np.sin(2*np.pi*x/lambdaN)
        
                   
    elif tube=='closed':
        lambdaN = 4*L/N
        if np.mod(N,2)==0:
            yp = np.zeros(len(x))
        else:
            yp = A * np.cos(2*np.pi*x/lambdaN)
        
    yn = - yp
    
    fig = plt.figure()
    plt.plot(x, yp, color='k')
    plt.plot(x, yn, color='k', ls='dashed')
    plt.axis('equal')
    plt.axis('off')
    
    fname = './standing_waves_' + tube + '_' + str(N) + '.pdf'
    plt.savefig(fname, bbox_inches='tight')

    return
```

```python
tube = 'open'
for N in range(1,6):
    standing_waves(N,tube)
```

```python
tube = 'closed'
for N in range(1,6):
    standing_waves(N,tube)
```
