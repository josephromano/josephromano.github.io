```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

from __future__ import division
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.ticker import NullFormatter
matplotlib.rcParams['savefig.dpi'] = 1.5 * matplotlib.rcParams['savefig.dpi']

import numpy as np
```

```python
ligo = np.loadtxt('aLIGO_ZeroDet_HighP.txt')
virgo = np.loadtxt('AdVirgo_noise.txt')
kagra = np.loadtxt('KAGRA_VRSED.txt')
```

## Read-in curves are in units of Hz^{-1/2}

```python
ligo[:,1] = np.sqrt(ligo[:,1]**2 * ligo[:,0])
virgo[:,1] = np.sqrt(virgo[:,1]**2 * virgo[:,0])
kagra[:,1] = np.sqrt(kagra[:,1]**2 * kagra[:,0])
```

```python
plt.plot(ligo[:,0],ligo[:,1],label='Advanced LIGO (zero det., high power)')
plt.plot(virgo[:,0],virgo[:,1],label='Advanced Virgo')
plt.plot(kagra[:,0],kagra[:,1],label='KAGRA (VRSED)')

plt.xscale('log')
plt.yscale('log')
plt.xlim(10.,10000.)
plt.ylim(1e-23,1e-19)

plt.xlabel('Frequency [Hz]')
plt.ylabel(r'$h_n(f) = \sqrt{f\times S_n(f)}$')

plt.legend(frameon=True, shadow=True, loc='upper center')
plt.savefig('advanced_noise_curves.pdf')
```
